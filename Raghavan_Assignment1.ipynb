```python
#3pt 12 



def return_random_sample_joint():
    randnum = random(1)
    if(randnum<=0.1):
      x,y = 0,1
    elif(randnum>0.1 and randnum<=0.3):
      x,y = 1,0
    elif(randnum>0.3 and randnum<=0.6):
      x,y = 1,1
    else:
      x,y = 0,0
    return x,y
print("Example outputs of the sampling function:")
print(return_random_sample_joint())
print(return_random_sample_joint())
print(return_random_sample_joint())
```

#### d (1.5 points)
Mathematically, the PDF of a sum of two random variables is the convolution of their individual PDFs (do you understand why?). Let's call their PDFs $f(x)$ and $g(x)$, then the convolution is defined as

$$ (f \circledast g)(x) = \int_{-\infty}^{\infty} f(\tau) g(x - \tau) d\tau \, .$$

Program this convolution yourself for the two normal distributions $\mathcal{N}(x|\mu=2, \sigma=1)$ and $\mathcal{N}(x|\mu=4, \sigma=2)$ and show that you can recover the same connections seen already for the sets $\vec{x}$, $\vec{y}$ and $\vec{z}$.

**Hint**: You are **not** allowed to use a predefined convolution function. Approximate the integral by a sum over small elements with width $\Delta \tau$.

```python
from scipy.stats import norm
def convolution_sum(f, g, x, delta_tau):
    '''
    Performs convolution operation on two pdfs.

    Parameters:
    -----------
    1. f- first pdf
    2. g- second pdf
    3. x- range of values over which the convolution is being evaluated
    4. delta_tau- width of small element
    Returns:
    --------
    Convolution sum of 2 pdfs for the input parameters.
    '''
    tau_values = np.arange(-10, 10, delta_tau)
    result = 0
    for tau in tau_values:
        result += f(tau) * g(x - tau) * delta_tau
    return result

f = lambda x: norm.pdf(x, 2, 1)
g = lambda x: norm.pdf(x, 4, 2)
x_values = np.linspace(1, 11, 100)
delta_tau = 0.1
convolution_result = [convolution_sum(f, g, x, delta_tau) for x in x_values]
plt.plot(x_values, convolution_result)
plt.title('Convolution of two normal distributions')
plt.xlabel('x')
plt.ylabel('PDF')
plt.show()

print('Mean of convolution of two normal distributions:',x_values[np.argmax(convolution_result)])
```

### 2.2 Marginal normal distributions (3 points)

*Topics: multivariate normal & marginal distributions*

Consider a multivariate Gaussian of the form
$$P(\mathbf{x}) = \mathcal{N}(\mathbf{x}|\mathbf{\mu}, \mathbf{\Sigma}) = \frac{1}{(2\pi)^{D/2} |\mathbf{\Sigma}|^{1/2}} \exp\left(-\frac{1}{2} (\mathbf{x}-\boldsymbol{\mu})^{T} \mathbf{\Sigma}^{-1}(\mathbf{x}-\boldsymbol{\mu})\right) \, ,$$

where $\mathbf{x}$ and $\boldsymbol{\mu}$ are vectors with $D$ entries and $\mathbf{\Sigma}$ is a $D \times D$ covariance matrix. $|\mathbf{\Sigma}|$ is the determinant of this matrix.

#### a (2 points)
Consider the case where $D = 2$. Use the following three sets of ($\boldsymbol{\mu}$, $\mathbf{\Sigma}$) to create 2D contour- and 3D surface plots of the corresponding PDFs (with arbitrary contour levels)

$$ \boldsymbol{\mu_1} = \begin{pmatrix} 5 \\ 5 \end{pmatrix} ~ \mathbf{\Sigma_1} = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} \, ,$$

$$ \boldsymbol{\mu_2} = \begin{pmatrix} 5 \\ 5 \end{pmatrix} ~ \mathbf{\Sigma_2} = \begin{pmatrix} 1 & -0.5 \\ -0.5 & 1 \end{pmatrix} \, ,$$

$$ \boldsymbol{\mu_3} = \begin{pmatrix} 5 \\ 5 \end{pmatrix} ~ \mathbf{\Sigma_3} = \begin{pmatrix} 1 & 0.7 \\ 0.7 & 1 \end{pmatrix} \, .$$

Implement the multivariate Gaussian yourself using numpy matrix multiplication.

```python
#3pt 18.5 



import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def multivariate_gaussian_pdf(x, mean, cov):
    '''
    Returns the pdf of a multivariate gaussian.

    Parameters:
    -----------
    1. x- 2x1 matrix with position at which the pdf is being evaluated
    2. mean- 2x1 matrix with mean of each normal distribution
    3. cov- 2x2 covariance matrix

    Returns:
    --------
    An array with probability density for different values of x
    '''
    const_term = 1 / ((2 * np.pi) * np.sqrt(np.linalg.det(cov)))
    exp_term = np.exp(-0.5 * np.dot(np.dot((x - mean).T, np.linalg.inv(cov)), (x - mean)))
    pdf = const_term * exp_term
    return pdf

x = np.linspace(2, 8, 100)
y = np.linspace(2, 8, 100)
X, Y = np.meshgrid(x, y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X
pos[:, :, 1] = Y

mean = np.array([5, 5])
cov1 = np.array([[1, 0], [0, 1]])
cov2 = np.array([[1, -0.5], [-0.5, 1]])
cov3 = np.array([[1, 0.7], [0.7, 1]])

Z1 = np.array([multivariate_gaussian_pdf(point, mean, cov1) for point in pos.reshape(-1, 2)])
Z1 = Z1.reshape(X.shape)

plt.figure(figsize=(6, 5))
plt.contour(X, Y, Z1, cmap='viridis')
plt.title('2D Contour Plot of Multivariate Gaussian with First Covariance Matrix',fontsize=10)
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar()
plt.show()
print('\n')
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z1, cmap='viridis')
ax.set_title('3D Surface Plot of Multivariate Gaussian')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('PDF')
plt.show()
print('\n')

Z2 = np.array([multivariate_gaussian_pdf(point, mean, cov2) for point in pos.reshape(-1, 2)])
Z2 = Z2.reshape(X.shape)

plt.figure(figsize=(6, 5))
plt.contour(X, Y, Z2, cmap='viridis')
plt.title('2D Contour Plot of Multivariate Gaussian with Second Covariance Matrix',fontsize=10)
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar()
plt.show()
print('\n')
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z2, cmap='viridis')
ax.set_title('3D Surface Plot of Multivariate Gaussian')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('PDF')
plt.show()
print('\n')

Z3 = np.array([multivariate_gaussian_pdf(point, mean, cov3) for point in pos.reshape(-1, 2)])
Z3 = Z3.reshape(X.shape)

plt.figure(figsize=(6, 5))
plt.contour(X, Y, Z3, cmap='viridis')
plt.title('2D Contour Plot of Multivariate Gaussian with Third Covariance Matrix',fontsize=10)
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar()
plt.show()
print('\n')
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z3, cmap='viridis')
ax.set_title('3D Surface Plot of Multivariate Gaussian')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('PDF')
plt.show()
```

To obtain a marginal of a multivariate normal distribution, we only need to remove the the variable that is being marginalized out, the resulting marginal is still a normal distribution.
To demonstrate this, I will be marginalizing $x_1$ out of the first multivariate gaussian.

$$ P(x_2) = \int_{-}^{}\frac{1}{2\pi \sqrt\Sigma}exp(-\frac{1}{2}\begin{pmatrix}x_1-5 & x_2-5 \end{pmatrix} \begin{pmatrix} 1 & 0 \\0 & 1 \end{pmatrix} \begin{pmatrix}x_1-5 \\ x_2-5 \end{pmatrix} dx_1$$
This simplifies to
$$P(x_2) = \int_{-}^{}\frac{1}{2\pi \sqrt\Sigma}exp(-\frac{1}{2}(x_1-5)^2)exp(-\frac{1}{2}(x_2-5)^2)dx_1 $$
$$P(x_2) = \frac{exp(-\frac{1}{2}(x_1-5)^2)}{2\pi \sqrt\Sigma}\int_{-}^{}exp(-\frac{1}{2}(x_1-5)^2)dx_1$$

The integral is a constant, so $P(x_2)$ is the normal distribution for $x_2$ times a constant.

We can make similar arguments for the other five marginals as only the constants outside and within the exponential will change in each case. The two variables $x_1$ and $x_2$ are independent and will always separate out such that the resulting marginal is a normal distribution.

```python
from scipy.stats.contingency import margins
marginal_dist1 = margins(Z1)
plt.plot(x,marginal_dist1[0])
plt.xlabel('x')
plt.ylabel('PDF')
plt.title('One Marginal distribution for first multivariate gaussian')
plt.show()

marginal_dist2 = margins(Z2)
plt.plot(x,marginal_dist2[0])
plt.xlabel('x')
plt.ylabel('PDF')
plt.title('One Marginal distribution for second multivariate gaussian')
plt.show()

marginal_dist3 = margins(Z3)
plt.plot(x,marginal_dist3[0])
plt.xlabel('x')
plt.ylabel('PDF')
plt.title('One Marginal distribution for third multivariate gaussian')
plt.show()
```
